###### replace coloumn names

```python
df.rename(columns={"Company": "FIRM"}, inplace=True)
df.rename(columns={"Age": "AGE"}, inplace=True)
df.rename(columns={"Salary": "INCOME"}, inplace=True)
df.rename(columns={"Place": "PLACE"}, inplace=True)
df.rename(columns={"Country": "COUNTRY"}, inplace=True)
df.rename(columns={"Gender": "GENDER"}, inplace=True)
df
```

# Data Cleaning

#### Handling duplicate

```python
df[df.duplicated()]
```

```python
# remove four duplicate raws

df=df.drop_duplicates()
df
```

#### Handling missing data

```python
df.isnull().sum()
```

```python
#replace 0 value with Nan

df["AGE"].replace(0,np.nan,inplace=True)
df["INCOME"].replace(0,np.nan,inplace=True)
df["FIRM"].replace(0,np.nan,inplace=True)
df["PLACE"].replace(0,np.nan,inplace=True)

df
```

#### replace nan values with median

```python
df['AGE']=df['AGE'].fillna(df['AGE'].median())

df['INCOME']=df['INCOME'].fillna(df['INCOME'].median())
#replacing nan value for feature places with others
df["PLACE"]=df["PLACE"].fillna("others")
#replacing nan value for feature company with others
df["FIRM"]=df["FIRM"].fillna("others")
df
```

##### outlier detection and removal

```python
#outlier detection and removal using zscore

df['zscore']=(df.AGE-df.AGE.mean())/df.AGE.std()
df[(df.zscore>-3)&(df.zscore<3)]

df['zscore']=(df.INCOME-df.INCOME.mean())/df.INCOME.std()
df[(df.zscore>-3)&(df.zscore<3)]
```

```python
#outlier detection and removal using IQR
m=df.AGE.mean()
s=df.AGE.std()
print(m-3*s)
print(m+3*s)

mn=df.INCOME.mean()
sd=df.INCOME.std()
print(mn-3*sd)
print(mn+3*sd)
#remove outlier
df_no_outliers=df[(df.AGE>7.46)&(df.AGE<56.92)]
df_no_outliers

df_no_out=df[(df.INCOME>-1873.73)&(df.INCOME<12350.11)]
df_no_out
```

### Data encoding

```python
#creating dummy columns for firm categories
dummies=pd.get_dummies(df.FIRM,dtype=np.int32)
dummies

#merging with company_merged df
encoded_company_merged = pd.concat([df,dummies],axis='columns')
encoded_company_merged.head()
```

```python
#Dropping the original Company column and one of the dummies so as to avoid dummy trap
final_company_encoded = encoded_company_merged.drop(['FIRM','others'], axis='columns')
final_company_encoded
```

```python
#creating dummy columns for place categories
place_dummies = pd.get_dummies(final_company_encoded.PLACE,dtype=np.int32)
place_dummies

#merging with company_merged df
encoded_place_merged = pd.concat([final_company_encoded,place_dummies],axis='columns')
encoded_place_merged.head()
```

```python
#Dropping the original Place column and one of the dummies so as to avoid dummy trap
# Also dropping Country column as it has no impact
final_encoded_df = encoded_place_merged.drop(['PLACE','others','COUNTRY'], axis='columns')
final_encoded_df
```

# feature scaling

```python
from sklearn.preprocessing import MinMaxScaler
 
#Copying the columns to scale to a new df
df_to_scale = df[['AGE','INCOME']]
df_to_scale
```

```python
scaler = MinMaxScaler()
#scaling the values to be between 0 & 1 using MinMax Scaling
scaled_data = scaler.fit_transform(df_to_scale)
scaled_df = pd.DataFrame(scaled_data,columns=df_to_scale.columns)
scaled_df.head()
```

```python
#dropping the original Age & Salary columns
final_encoded=final_encoded_df.drop(['AGE','INCOME'],axis='columns')

#Merging the scaled columns to final encoded dataframe
final_df = pd.concat([scaled_df,final_encoded],axis='columns')
final_df.head()
```
